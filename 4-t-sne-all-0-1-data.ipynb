```python
import pandas as pd
import numpy as np
from kmodes.kmodes import KModes
import plotly.graph_objects as go
from sklearn.manifold import TSNE
from sklearn.externals import joblib
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# PCA for data visualization
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', -1)  # or 199
```

```python
data_file = r"Data\Cleaned\numeric_data"       
labels_file = r"Data\Cleaned\labels"       
data = joblib.load(data_file)
labels_df = joblib.load(labels_file)
```

```python
X = data
labels_list = labels_df['labels'].value_counts()
labels_list
```

tsne_model = TSNE(n_components=2, random_state = 0, perplexity = 50, 
                  n_iter = 5000).fit_transform(X)   



```python
tsne_data = np.vstack( ( tsne_model.T, labels_df['labels'].values ) ).T
tsne_df = pd.DataFrame( data= tsne_data, columns= ("Dim1", "Dim2", "labels" ) )

tsne_df = pd.DataFrame( data= tsne_data, columns= ("Dim1", "Dim2", "labels" ) )


plt.figure(figsize=(16,7))
sns.scatterplot(
    x="Dim1", y="Dim2",
    hue="labels",
    palette=sns.color_palette("hls", 6),
    data=tsne_df,
    legend="full",
    alpha=0.3
)
```

# TSNE on PCA

```python
pca_X = PCA(n_components=50).fit_transform(X)
tsne_model = TSNE(n_components=2, random_state = 0, perplexity = 50, 
                  n_iter = 5000).fit_transform(pca_X)   


tsne_data = np.vstack( ( tsne_model.T, labels_df['labels'].values ) ).T
tsne_df = pd.DataFrame( data= tsne_data, columns= ("Dim1", "Dim2", "labels" ) )

tsne_df = pd.DataFrame( data= tsne_data, columns= ("Dim1", "Dim2", "labels" ) )


plt.figure(figsize=(16,7))
sns.scatterplot(
    x="Dim1", y="Dim2",
    hue="labels",
    palette=sns.color_palette("hls", 6),
    data=tsne_df,
    legend="full",
    alpha=0.3
)
```

```python
pca_X = PCA(n_components=3).fit_transform(X)
tsne_model = TSNE(n_components=2, random_state = 0, perplexity = 50, 
                  n_iter = 5000).fit_transform(pca_X)   


tsne_data = np.vstack( ( tsne_model.T, labels_df['labels'].values ) ).T
tsne_df = pd.DataFrame( data= tsne_data, columns= ("Dim1", "Dim2", "labels" ) )

tsne_df = pd.DataFrame( data= tsne_data, columns= ("Dim1", "Dim2", "labels" ) )


plt.figure(figsize=(16,7))
sns.scatterplot(
    x="Dim1", y="Dim2",
    hue="labels",
    palette=sns.color_palette("hls", 6),
    data=tsne_df,
    legend="full",
    alpha=0.3
)
```
