```python
filepath = r"Data\Cleaned\for-analysis.csv"       
# load data
df = joblib.load(filepath)
```

```python
# remove first row - questions columnen
df = df.iloc[ 1: , : ]
print(df.shape)
print(df.head(5))
```

```python
print("old classes")
print(df['Q6'].value_counts())
print("New classes = ")
print(df['labels'].value_counts())
```

```python
# drop Q6
df.drop( ['Q6'],axis = 1, inplace = True )
```

```python
print(df.columns)
# print shape 
print(df.shape)
# Check the null values
print(df.isnull().sum().sort_values(ascending = False))
```

```python
# store all column names into a list
all_cols = df1.columns.tolist()

print(all_cols)
# remove Q6 - lables from the all_cols and filter parts and not parts cols
all_cols.remove('labels')

not_part_col = []
part_col = []

for col in all_cols:
    if "Part" in col:
        part_col.append(col)
    else:
        not_part_col.append(col)
        
print("COlumns with parts = ")
print(part_col)
print("Columns without parts = ")
print(not_part_col)
    
```

# df_parts Conversion into 0 and 1

```python
print("Details of part df = ")
df_parts = df1[part_col]
print(df_parts.shape)
#print(df_parts.head(5))
df_parts.head(5)
```

```python
# change all blank fields to 0 and string fields to 1
df_parts = df_parts.replace(np.nan, '0', regex=True)

def check_str_len(x):
    if len(x) > 1:
        return 1.0
    else:
        return 0.0
    
for col in part_col:
    print("BEFORE converting 0 and 1 :  "+col)
    #df_parts[[col] = df_parts[[col].applymap(str)
    print(df_parts[col].head(10))
    df_parts[col] = df_parts[col].apply( lambda x : check_str_len(str(x)) )
    print("AFTER converting 0 and 1 :  "+col)
    print(df_parts[col].head(10))
```

```python
df_parts.head(10)
```

# df_not_parts conversion into numeric

```python
print("Details of NOT part df = ")
df_not_parts = df1[not_part_col]
print(df_not_parts.shape)
print(df_not_parts.head(5))
```

```python
# convert columns into one-hot encoding
for col in not_part_col[1:]:
    print("Processing...     "+col)
    # Get one hot encoding of columns columns
    one_hot = pd.get_dummies(df_not_parts[col])
    one_hot = (pd.get_dummies(df_not_parts[col], prefix='', prefix_sep='')
                   .max(level=0, axis=1)
                   .add_prefix(col+'-'))  
    # Drop column that column as it is now encoded
    df_not_parts = df_not_parts.drop(col,axis = 1)
    # Join the encoded df
    df_not_parts = df_not_parts.join(one_hot)
    print( "Columns = " )
    print(df_not_parts.columns)

```
